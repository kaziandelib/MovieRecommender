```python
# Import the 'ast' module for safely evaluating string expressions into Python objects
# This is often used to convert string representations of lists or dictionaries back into their original types
import ast
```

```python
# Function to extract 'name' fields from a stringified list of dictionaries and remove spaces from names
def extract_names(x):
    try:
        # Parse the string to a Python list of dicts and extract 'name', removing spaces
        return [i['name'].replace(" ", "") for i in ast.literal_eval(x)]
    except:
        # Return empty list if parsing fails or data is malformed
        return []

# Function to extract the Director's name from the 'crew' column (list of dicts)
def extract_director(x):
    try:
        crew_list = ast.literal_eval(x)  # Convert stringified list to Python list
        # Return a list with director's name(s), removing spaces
        return [i['name'].replace(" ", "") for i in crew_list if i['job'] == 'Director']
    except:
        return []

# Function to extract top 5 cast members' names, removing spaces
def extract_top_cast(x):
    try:
        # Convert string to list, take first 5 items, extract 'name', remove spaces
        return [i['name'].replace(" ", "") for i in ast.literal_eval(x)[:5]]
    except:
        return []

# Apply the above extraction functions to relevant columns in the movies DataFrame
movies['genres'] = movies['genres'].apply(extract_names)  # Extract genres as cleaned list of names
movies['keywords'] = movies['keywords'].apply(extract_names)  # Extract keywords similarly
movies['cast'] = movies['cast'].apply(extract_top_cast)  # Extract top 5 cast members
movies['crew'] = movies['crew'].apply(extract_director)  # Extract director(s)
movies['production_companies'] = movies['production_companies'].apply(extract_names)  # Extract production company names
```

```python
# Display the first few rows of the movies DataFrame after cleaning and extracting features
movies.head()
```

```python
# Convert the 'release_date' column to datetime format for easier date manipulation
movies['release_date'] = pd.to_datetime(movies['release_date'])

# Extract the year from 'release_date' and create a new column 'release_year'
movies['release_year'] = movies['release_date'].dt.year

# Display 2 random samples from the movies DataFrame to verify changes
movies.sample(2)
```

# Building the recommender

```python
# Create a new 'tags' column by combining multiple text-based features into a single string for each movie
movies['tags'] = (
    movies['overview'] + ' ' +  # Movie plot summary text
    movies['genres'].apply(lambda x: ' '.join(x)) + ' ' +  # Join genres list into a single string
    movies['keywords'].apply(lambda x: ' '.join(x)) + ' ' +  # Join keywords list into a string
    movies['cast'].apply(lambda x: ' '.join(x)) + ' ' +  # Join top cast members' names into a string
    movies['crew'].apply(lambda x: ' '.join(x)) + ' ' +  # Join director's name(s) into a string
    movies['production_companies'].apply(lambda x: ' '.join(x))  # Join production companies' names into a string
)
# This combined 'tags' field will be useful for text-based similarity or recommendation algorithms
```

```python
# Display 2 random samples from the movies DataFrame to inspect the newly created 'tags' column and other data
movies.sample(2)
```

```python
# Select only the 'movie_id', 'title', and the combined 'tags' columns for further analysis or modeling
use_data = movies[['movie_id', 'title', 'tags']]

# Display the first 5 rows of the selected data to verify the selection
use_data.head()
```

```python
# Import CountVectorizer from sklearn to convert text data into a matrix of token counts
from sklearn.feature_extraction.text import CountVectorizer

# Initialize CountVectorizer with a max feature limit of 3000 and removing English stop words
cv = CountVectorizer(max_features=3000, stop_words='english')

# Display the CountVectorizer object to verify initialization
cv
```

```python
# Fit the CountVectorizer on the 'tags' column and transform text data into a numerical feature matrix
# Use .values.astype('U') to ensure the text data is unicode (required by CountVectorizer)
vec = cv.fit_transform(use_data['tags'].values.astype('U')).toarray()

# Display the resulting array of token counts for each movie
vec
```

```python
# Check the shape of the feature matrix
# Rows correspond to movies, columns correspond to the 3000 most frequent words (features)
vec.shape
```

```python
# Compute cosine similarity matrix to measure pairwise similarity between movie tag vectors
# Each value sim[i][j] represents how similar movie i is to movie j (1 = identical, 0 = completely different)
from sklearn.metrics.pairwise import cosine_similarity
sim = cosine_similarity(vec)

sim  # Display the similarity matrix
```

```python
# Retrieve the row(s) corresponding to the movie titled 'The Dark Knight'
use_data[use_data['title'] == 'The Dark Knight']
```

```python
# Sort movies by similarity score to the movie at index 65 (likely 'The Dark Knight')
# enumerate(sim[65]) pairs each movie's index with its similarity score to movie 65
# sorted by similarity score in descending order (most similar first)
distance = sorted(list(enumerate(sim[65])), reverse=True, key=lambda vec: vec[1])

# Display sorted list of tuples: (movie_index, similarity_score)
distance
```

```python
# Print titles of the top 5 most similar movies based on similarity scores
for i in distance[0:5]:
    print(use_data.iloc[i[0]].title)
```

# Testing the recommender

```python
def recommender(movie_title):
    # Find the index of the movie in use_data by title
    index = use_data[use_data['title'] == movie_title].index[0]
    
    # Get similarity scores for this movie with all others and sort descending
    distance = sorted(list(enumerate(sim[index])), reverse=True, key=lambda x: x[1])
    
    # Print the titles of the top 5 most similar movies (excluding the movie itself)
    for i in distance[1:6]:
        print(use_data.iloc[i[0]].title)
```

```python
# Call the function with the movie title 'The Dark Knight Rises'
recommender('The Dark Knight Rises')
```
